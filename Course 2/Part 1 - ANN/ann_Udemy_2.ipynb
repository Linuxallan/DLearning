# Construir la ANN
## Importar Keras y librerias adicionales

```python
import keras 
from keras.models import Sequential
from keras.layers import Dense
```

## Inicializar la ANN
### El modelo va a Clasificar.

```python
classifier = Sequential()
```

## Añadir capa de entrada, capas ocultas y capa de salida

```python
# Para conocer la cantidad de 'Input Features' hago lo siguiente
len(X_train[0])
```

```python
# Dense : define la sinapsis (cantidad de 'Input Features' que recibira la 'Hidden Layer'), y define la 'Hidden Layer' y sus 'Output Features'.
# Units : por recomendacion del profesor se escojen la cantidad de neuronas que corresponden a la media de 'Input Features'.
# Input_Dim : definimos la cantidad de 'Input Features' que recibira la 'Hidden Layer'.
# Kernel:Inicializer : definimos los valores de inicializacion de los parametros de la ANN.
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))

# 'Hidden Layer' 2
# 'Input_Dim' no es necesario especificarla (no especificar la cantidad de 'Input Features')
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

# 'Output Layer'
# Modelo de Clasificacion (dos clases de salida).
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

## Compilar la ANN

```python
# optimizar : Optimization Function como Stochastic Greadient Descent.
# loss : Error/Costo Function.
# metrics : metrica de para ir visualizando el rendimiento del entrenamiento del modelo.
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

# Training del modelo de ANN

```python
# batch_size : contenedor/porcion de datos del conjunto de datos de entrenamiento de 32 datos.
# --> ingesta de paquetes de 32 datos a la ANN para que se entrene.
# epochs : numero de ingesta completa del conjunto de datos de entrenamiento a la ANN.
classifier.fit(X_train, y_train, batch_size = 32, epochs = 100)
```

## Prediccion y evaluacion de la ANN

```python
y_pred = classifier.predict(X_test)
```

```python
# Trnasformar los datos de prediccion a booleanos.
# --> los verdaderos son clientes que se van del banco.
# --> los falsos son clientes que se quedan en el banco.
print(y_pred)
print(y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

# Parte 2 - Evaluar, Mejorar y ajustar la RNA

## Evaluar la RNA

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.layers import Dropout
```

### Crear el Modelo.
### Dropuot : es para evitar el Overfitting apagando un porcentje de neuronas en cada capa.

```python
def build_classifier():

  # Inicializar la ANN
  classifier = Sequential()

  # Añadir capa de entrada y primera capa oculta
  classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))

  # Capa Dropuot (apagar neuronas para evitar el Overfitting) a la primera 'Hidden Layer'
  # p = porcentaje de apagado de neuronas en la capa.
  classifier.add(Dropout(0.5))

  # Añadir segunda capa oculta
  classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

  # Capa Dropuot (apagar neuronas para evitar el Overfitting) a la segunda 'Hidden Layer'
  classifier.add(Dropout(0.5))

  # Añadir capa de salida
  classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

  # Compilar la ANN
  classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

  # Devolder el modelo de calsificacion
  return classifier
```

```python
classifier = KerasClassifier(build_fn = build_classifier, batch_size = 32, nb_epoch = 10)
# cv : cross validations --> de decimos de 32, lo que significa que tomara lotes de 32 datos y
# dividira entre entrenamiento y testing, variandolos en cada iteracion para que todos los datos
# se usen tanto en Training como en Testing (K-Fold Cross Validation)
# n_jobs : le decimos que use todos los recursos posibles que pueda tener del hardware (pc)
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 32, verbose = 1)
```

```python
accuracies
```

```python
mean = accuracies.mean()
variance = accuracies.std() # Una varianza sobre 0,1 es alta.
print(mean)
print(variance)
```

# Ajustar la ANN

```python
# Esta funcion prueba todas las combinaciones posibles que pueden tener los hiperparametros
# del modelo, lo hace para encontrar el mejor modelo posible.
from sklearn.model_selection import GridSearchCV
```

### Creacion del Modelo

```python
def build_classifier(optimizer):

  # Inicializar la ANN
  classifier = Sequential()

  # Añadir capa de entrada y primera capa oculta
  classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))

  # Capa Dropuot (apagar neuronas para evitar el Overfitting) a la primera 'Hidden Layer'
  # p = porcentaje de apagado de neuronas en la capa.
  # classifier.add(Dropout(p = 0.1))

  # Añadir segunda capa oculta
  classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

  # Capa Dropuot (apagar neuronas para evitar el Overfitting) a la segunda 'Hidden Layer'
  # classifier.add(Dropout(p = 0.1))

  # Añadir capa de salida
  classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

  # Compilar la ANN
  classifier.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])

  # Devolder el modelo de calsificacion
  return classifier
```

```python
classifier = KerasClassifier(build_fn = build_classifier)
```

```python
parameters = {
    'batch_size': [32, 64, 128],
    'nb_epoch': [50, 100, 200],
    'optimizer': ['adam', 'rmsprop']
}
```

```python
# Este metodo de ajuste para encontrar el mejor modelo no permite el 'Dropout' no se por que.
grid_search = GridSearchCV(estimator = classifier, param_grid = parameters, scoring = 'accuracy', cv = 10, n_jobs = -1)
```

```python
grid_search = grid_search.fit(X_train, y_train)
best_parameters = grid_search.best_params_
best_accuracy = grid_search.best_score_
```
