# Pre procesado de datos
### Codificar datos Categoricos

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
```

# Dividir el DataSet en conjunto de Training y Testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

# Escalado de variables

```python
X_train[0]
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
X_train[0]
```

```python
X_test[0]
```

# Construir la ANN
## Importar Keras y librerias adicionales

```python
import keras 
from keras.models import Sequential
from keras.layers import Dense
```

## Inicializar la ANN
### El modelo va a Clasificar.

```python
classifier = Sequential()
```

## AÃ±adir capa de entrada, capas ocultas y capa de salida

```python
# Para conocer la cantidad de 'Input Features' hago lo siguiente
len(X_train[0])
```

```python
# Dense : define la sinapsis (cantidad de 'Input Features' que recibira la 'Hidden Layer'), y define la 'Hidden Layer' y sus 'Output Features'.
# Units : por recomendacion del profesor se escojen la cantidad de neuronas que corresponden a la media de 'Input Features'.
# Input_Dim : definimos la cantidad de 'Input Features' que recibira la 'Hidden Layer'.
# Kernel:Inicializer : definimos los valores de inicializacion de los parametros de la ANN.
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))

# 'Hidden Layer' 2
# 'Input_Dim' no es necesario especificarla (no especificar la cantidad de 'Input Features')
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

# 'Output Layer'
# Modelo de Clasificacion (dos clases de salida).
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

## Compilar la ANN

```python
# optimizar : Optimization Function como Stochastic Greadient Descent.
# loss : Error/Costo Function.
# metrics : metrica de para ir visualizando el rendimiento del entrenamiento del modelo.
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

# Training del modelo de ANN

```python
# batch_size : contenedor/porcion de datos del conjunto de datos de entrenamiento de 32 datos.
# --> ingesta de paquetes de 32 datos a la ANN para que se entrene.
# epochs : numero de ingesta completa del conjunto de datos de entrenamiento a la ANN.
classifier.fit(X_train, y_train, batch_size = 32, epochs = 100)
```

## Prediccion y evaluacion de la ANN

```python
y_pred = classifier.predict(X_test)
```

```python
# Trnasformar los datos de prediccion a booleanos.
# --> los verdaderos son clientes que se van del banco.
# --> los falsos son clientes que se quedan en el banco.
print(y_pred)
print(y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
