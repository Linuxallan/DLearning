## Librerias

```python
import tensorflow as tf
tf.keras.backend.clear_session()
print(tf.__version__)
```

```python
inputs = tf.keras.Input(shape=(784, ))
x = tf.keras.layers.Dense(units=32, activation='relu', name='layer_1')(inputs)
x = tf.keras.layers.Dense(units=16, activation='relu', name='layer_2')(x)
outputs = tf.keras.layers.Dense(units=1, activation='sigmoid')(x)
```

```python
model = tf.keras.Model(inputs=inputs, outputs=outputs, name='model_1')
```
